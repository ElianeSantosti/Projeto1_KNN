```python
# Definindo a função de cálculo da distância euclidiana
# A função calcula a distancia entre as compras de cada cliente em cliente1 e clientes2.
def DistEuclidiana(clientes1, clientes2):
    lista_final = []
    for elemento1 in range(len(clientes1)):
        compras_nao_classificadas = clientes1[elemento1][2]
        distancias = []
        for elemento2 in range(len(clientes2)):
            compras_classificadas = clientes2[elemento2][2]
            distancia_calculada = sum((x - y) ** 2 for x, y in zip(compras_nao_classificadas, compras_classificadas)) ** 0.5 #elevado ao quadrado
            '''
            A expressao calcula a distancia euclidiana entre os pontos no espaço multidimensional.
            É uma metrica usada para medir a proximidade entre os pontos de classificação
            Será usado para determinar quais são os vizinhos mais proximos.
            '''
            distancias.append((distancia_calculada, clientes2[elemento2][1], compras_classificadas))
        lista_final.append((clientes1[elemento1][0], distancias))
    return lista_final

def KNN(nps_classificada):
    classe = []
    for cpf_cliente, distancias in nps_classificada:
        promotor, neutro, detrator = 0, 0, 0
        for distancia, nps_cliente, ultimas_compras in distancias:
            if nps_cliente == 'Promotor':
                promotor += 1
            elif nps_cliente == 'Neutro':
                neutro += 1
            else:
                detrator += 1

        freq = max(promotor, neutro, detrator)

        if freq == promotor:
            nps = 'Promotor'
        elif freq == neutro:
            nps = 'Neutro'
        else:
            nps = 'Detrator'

        classe.append([cpf_cliente, nps, ultimas_compras])

    return classe

    # Definindo a função de ordenação das distâncias
'''
Esse código em conjunto executa o processo de classificação dos clientes não classificados
com base nos vizinhos mais próximos e imprime o CPF e a classificação correspondente.
'''

def Ordenacao(clientes, K):
    dados_ordenados = []
    for dist in clientes:
        cresc = sorted(dist[1])[:K]
        dados_ordenados.append([dist[0], cresc])
    return dados_ordenados

nps_classificada = DistEuclidiana(no_class, data)

k = int(input("Digite o valor de K para o KNN: "))

dados_ordenados = Ordenacao(nps_classificada, k)

classificados = KNN(dados_ordenados)

print("Lista no_class classificada")

for cpf, nps, ultimas_compras in classificados:
    print(f'CPF: {cpf} \u2192 NPS: {nps} \u2192 Últimas compras: {ultimas_compras}')
```
